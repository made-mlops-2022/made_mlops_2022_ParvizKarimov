## Pandas profiling visualisation

```python
profile = ProfileReport(df, title="Dataset Profiling")
```

```python
from IPython.core.display import display, HTML
display(HTML(profile.to_html()))
```

### Вытащим чиселки для корреляции Спирмана

```python
correlations = profile.description_set["correlations"]
print(correlations.keys())
```

```python
correlations['spearman'].style.background_gradient(cmap='plasma').format(precision=2)
```
